```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
def fetch_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data

start_date = (datetime.today() - pd.DateOffset(years=5)).strftime('%Y-%m-%d')
end_date = datetime.today().strftime('%Y-%m-%d')

nifty_data = fetch_data('^NSEI', start_date, end_date)
```

# Nifty 

```python
nifty_data = nifty_data.dropna()
```

```python
nifty_data['Daily Return'] = nifty_data['Adj Close'].pct_change()
```

### Visualise Closing Prices 

```python
plt.figure(figsize=(14, 7))
plt.plot(nifty_data['Adj Close'], color='red', label='Nifty')
plt.title('Nifty Closing Prices (5 Years)')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend()
plt.show()
```

### Visualize Daily Returns

```python
plt.figure(figsize=(14, 7))
plt.plot(nifty_data['Daily Return'],  color='green', label='Nifty Daily Returns')
plt.title('Nifty Daily Returns (5 Years)')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```
